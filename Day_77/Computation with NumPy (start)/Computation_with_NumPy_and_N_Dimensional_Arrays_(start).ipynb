# Understanding NumPy's ndarray

NumPy's most amazing feature is the **powerful** ndarray.

<img src="https://i.imgur.com/1IUqnxX.png" width=200>

#### 1-Dimensional Arrays (Vectors)

```python
my_array = np.array([1.1,9.2,8.1,4.7]) # Create a 1D array
my_array.shape # Check the shape of the array
```

```python
print(my_array[2]) # Access the third element of the array
my_array.ndim # Check the number of dimensions of the array
```

#### 2-Dimensional Arrays (Matrices)

```python
array_2d = np.array([[1, 2, 3, 9], 
                     [5, 6, 7, 8]])

print(f'array_2d has {array_2d.ndim} dimensions')
print(f'Its shape is {array_2d.shape}')
print(f'It has {array_2d.shape[0]} rows and {array_2d.shape[1]} columns')
print(array_2d)
```

```python
print(array_2d[0, :]) # Access the first row
array_2d[1,2] # Access the element in the second row and third column
```

**Challenge**:

```python
mystery_array = np.array([[[0, 1, 2, 3],
                           [4, 5, 6, 7]],
                        
                         [[7, 86, 6, 98],
                          [5, 1, 0, 4]],
                          
                          [[5, 36, 32, 48],
                           [97, 0, 27, 18]]])


print(f'We have {mystery_array.ndim} dimensions')
print(f'The shape is {mystery_array.shape}') 
#Here, we see that the first dimension has 3 elements, the second has 2 elements, and the third has 4 elements
```

```python
mystery_array[2,1,:] # Access the second row of the third "block"
```

```python
mystery_array[:,:,0] # Access the first column of all "blocks"
```

# Linear Algebra with Vectors

```python
v1 = np.array([4, 5, 2, 7])
v2 = np.array([2, 1, 3, 3]) 
v1 + v2 # Element-wise addition of two arrays
```

```python
# Python Lists vs ndarrays
list1 = [4, 5, 2, 7]
list2 = [2, 1, 3, 3]
list1 + list2  # Concatenates the lists, not element-wise addition
```

```python
v1*v2 # Element-wise multiplication of two arrays
```

```python
#list1 * list2
# This will raise an error because Python lists do not support element-wise multiplication
```

# Broadcasting and Scalars


```python
array_2d = np.array([[1, 2, 3, 4], 
                     [5, 6, 7, 8]])
print(array_2d + 10) # Add 10 to each element of the array
print(array_2d * 2) # Multiply each element of the array by 2
print(array_2d - 5) # Subtract 5 from each element of the array
```

# Matrix Multiplication with @ and .matmul()

<img src=https://i.imgur.com/LCdhmi8.png width=350>

```python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
```

**Challenge**: Let's multiply `a1` with `b1`. Looking at the wikipedia example above, work out the values for c12 and c33 on paper. Then use the [`.matmul()`](https://numpy.org/doc/stable/reference/generated/numpy.matmul.html) function or the `@` operator to check your work. 

```python
c = np.matmul(a1, b1) # Matrix multiplication
print(f'Matrix c has {c.shape[0]} rows and {c.shape[1]} columns.')
c
```

```python
a1 @ b1 # Matrix multiplication using the @ operator
# This is equivalent to np.matmul(a1, b2) and is often more readable
```
